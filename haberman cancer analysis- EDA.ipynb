```python
# change the label of column 'status' ; 1:survived-'yes' , 2:not survived-'no'
data['status'] = data['status'].map({1:'yes',2:'no'})
data["status"].value_counts()
```

```python
print(data['status'].unique())
```

```python
# data points per class
print(data.groupby('status').count())
```

```python
data.describe()
```

```python
print('servived:yes ,  stats')
status_yes=data.loc[data['status']=='yes']
print(status_yes.describe())

print('\n')

print('servived:no, stats')
status_no=data.loc[data["status"]=="no"]
print(status_no.describe())
```

```python
# lets plot 2D in data
sns.set_style('whitegrid')
sns.scatterplot(data=data,x='nodes',y='age', hue='status')
plt.title('2-D scatter plot nodes vs age')
plt.show()
```

```python
# pair plots
sns.set_style('darkgrid')
sns.pairplot(data,hue='status')
plt.show()
```

```python
#PDF plot of ages

sns.set_style("dark");
sns.FacetGrid(data,hue="status",height=5)\
    .map(sns.distplot,"age")\
    .add_legend();

plt.title('Histogram of ages of patients', fontsize=17)
plt.show();
```

```python
# PDF plot of 'year'
sns.FacetGrid(data, hue="status", size=5) \
 .map(sns.distplot, "year") \
 .add_legend();
plt.title("Year of Operation histogram")
plt.show()
```

```python
# PDF plot of nodes
sns.FacetGrid(data, hue="status", size=4) \
 .map(sns.distplot, "nodes") \
 .add_legend();
plt.title("Positive Aux_Nodes Hitogram")
plt.show()
```

```python
sns.boxplot(x='status',y='nodes',hue="status", data=data)
plt.title("Box-plot for nodes")
plt.show()
```

```python
sns.violinplot(x='status',y='nodes',hue="status", data=data)
plt.title("Violin-plot for nodes")
plt.show()
```

```python
sns.catplot(x='status',y='nodes',hue="status", data=data,kind='boxen')
plt.show()
```
